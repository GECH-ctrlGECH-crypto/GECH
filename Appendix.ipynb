```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for demonstration
frames = ['Human Interest', 'Responsibility', 'Economic', 'Moral', 'Conflict']
al_jazeera = [44.4, 42.4, 10, 3, 27]
cnn = [40, 53, 6.6, 13.3, 22.2]

# Plotting the bar chart
x = np.arange(len(frames))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6))
bars1 = ax.bar(x - width/2, al_jazeera, width, label='Al Jazeera', color='blue')
bars2 = ax.bar(x + width/2, cnn, width, label='CNN', color='orange')

# Adding labels, title and custom x-axis tick labels
ax.set_xlabel('Frames')
ax.set_ylabel('Frequency')
ax.set_title('Comparison of News Frames Used by Al Jazeera and CNN')
ax.set_xticks(x)
ax.set_xticklabels(frames)
ax.legend()

# Adding the Chi-square statistic and p-value separately
plt.text(0.95, 0.85, 'Chi-square statistic: 6.5956', horizontalalignment='center',
         verticalalignment='center', transform=ax.transAxes, fontsize=12)
plt.text(0.95, 0.80, 'p-value: 0.1588', horizontalalignment='center',
         verticalalignment='center', transform=ax.transAxes, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
frames = ['Human interest', 'Responsibility frame', 'Economic', 'Moral', 'Conflict']
aljazeera = [44.4, 42.4, 10, 3, 27]
cnn = [40, 53, 6.6, 13.3, 22.2]

# Create bar chart
barWidth = 0.35
r1 = np.arange(len(frames))
r2 = [x + barWidth for x in r1]

plt.bar(r1, aljazeera, color='blue', width=barWidth, edgecolor='grey', label='Aljazeera')
plt.bar(r2, cnn, color='red', width=barWidth, edgecolor='grey', label='CNN')

# Add labels and title
plt.xlabel('Frames', fontweight='bold')
plt.ylabel('Frequency (%)', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(frames))], frames)
plt.legend()

# Add percentage labels on top of bars
for i in range(len(aljazeera)):
    plt.text(r1[i] - 0.1, aljazeera[i] + 0.5, f'{aljazeera[i]}%', color='black', fontweight='bold')
    plt.text(r2[i] - 0.1, cnn[i] + 0.5, f'{cnn[i]}%', color='black', fontweight='bold')

plt.title('Comparison of News Frames Used by Aljazeera and CNN')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Government', 'NGO/International Organization', 'Civilians']
aljazeera = [27, 56, 15]
cnn = [13, 21, 11]

# Colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Custom colors for better visualization

# Create a figure with a grid layout
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for Al Jazeera
axes[0].pie(aljazeera, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'grey'})
axes[0].set_title('Al Jazeera')

# Pie chart for CNN
axes[1].pie(cnn, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'grey'})
axes[1].set_title('CNN')

# Separate the charts visually by adding a grid-like background
for ax in axes:
    ax.set_facecolor('#f0f0f0')  # Light grey background to enhance grid visibility
    ax.grid(visible=True, linestyle='--', linewidth=0.5, color='grey')  # Dashed grid lines

# Overall title for the figure
fig.suptitle('Source Distribution Used by Al Jazeera and CNN', fontsize=16, fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Al Jazeera', 'CNN']
humanitarian_stories = [99, 45]
total_stories = [266, 99]
other_stories = [total - hum for total, hum in zip(total_stories, humanitarian_stories)]

# Plot
barWidth = 0.6
r = np.arange(len(categories))

# Create the bars
plt.bar(r, humanitarian_stories, color='lightgreen', edgecolor='grey', width=barWidth, label='Humanitarian Crisis Stories')
plt.bar(r, other_stories, bottom=humanitarian_stories, color='darkgreen', edgecolor='grey', width=barWidth, label='Other Stories')

# Add labels and title
plt.xlabel('News Outlets', fontweight='bold')
plt.ylabel('Number of Stories', fontweight='bold')
plt.xticks(r, categories)
plt.legend(loc='upper right')
plt.title('Proportion of Humanitarian Crisis Stories in Total News Coverage')

# Show the plot
plt.show()
```
